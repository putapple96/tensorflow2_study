## MNIST Neural Network

data set 을 적절하게 reshape 해준다

```python
import numpy as np
import tensorflow as tf
import random

random.seed(1234)
lr = 0.001
batch_size = 100
train_epochs = 15
classes = 10

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

print(x_train.shape)
```

```python
x_train = x_train.reshape(x_train.shape[0], x_train.shape[1] * x_train.shape[2])
x_test = x_test.reshape(x_test.shape[0], x_test.shape[1] * x_test.shape[2])

y_train = tf.keras.utils.to_categorical(y_train, classes)
y_test = tf.keras.utils.to_categorical(y_test, classes)
```

Neural Network model을 만들어 준다.  
input > relu - relu - softmax > output

```python
tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.Dense(input_dim=784, units=256, activation='relu'))
tf.model.add(tf.keras.layers.Dense(units=256, activation='relu'))
tf.model.add(tf.keras.layers.Dense(units=classes, activation='softmax'))

tf.model.compile(
    loss='categorical_crossentropy',
    optimizer=tf.keras.optimizers.Adam(lr=lr),
    metrics=['accuracy']
)

tf.model.summary()
```

```python
train_history = tf.model.fit(x_train, y_train, batch_size = batch_size, epochs=train_epochs)
```

```python
y_pred = tf.model.predict(x_test)
for x in range(0, 100):
    rand_idx = random.randint(0, x_test.shape[0]-1)
    print("index: ", rand_idx,
         "actual y: ", np.argmax(y_test[rand_idx]),
         "predicted y :", np.argmax(y_pred[rand_idx]))
    if(np.argmax(y_test[rand_idx]) ==  np.argmax(y_pred[rand_idx])):
        print("Correct!")
    else: print("Incorrect..")
```

```python
evaluate = tf.model.evaluate(x_test, y_test)
print('loss: ', evaluate[0])
print('accuracy: ', evaluate[1])
```
