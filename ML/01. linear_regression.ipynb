## Linear Regression

```python
import tensorflow as tf
import numpy as np
```

```python
x_train = [1, 2, 3, 4]
y_train = [0, -1, -2, -3]

# training data 
```

순차형(Sequential) 모델 생성 : 차례로 계층을 쌓아나가는 형태

```python
tf.model = tf.keras.Sequential()
```

계층을 쌓아준다. Dense 계층은 fully-connected layer를 의미함

```python
tf.model.add(tf.keras.layers.Dense(units=1, input_dim=1))
# units : output shape, imput_dim : input shape
```

Sequential 모델 정의후 학습 전 complie 메소드를 활용하여 설정을 해주어야함.
* Opitimizer : 최적화 함수
* Loss function : 손실 함수
* Metrics : 모델 성능 판정하는 지표 함수

이 외에도 몇가지 매개변수가 더 있음.

```python
sgd = tf.keras.optimizers.SGD(lr=0.1)

# learning rate를 0,1로한 standard GD

tf.model.compile(loss='mse', optimizer=sgd)
# mse = mean square error

tf.model.summary()
```

```python
tf.model.fit(x_train, y_train, epochs=200)
```

```python
y_predict = tf.model.predict(np.array([5,4]))
print(y_predict)
```
