## XOR_vanishing_gradient

vanishing gradient가 발생하도록 activation function 은 sigmoid로 설정해주고
레이어를 깊게 쌓아준다

```python
import tensorflow as tf
import numpy as np

x_data = np.array([[0, 0], [0, 1], [1, 0], [1, 1]], dtype=np.float32)
y_data = np.array([[0], [1], [1], [0]], dtype=np.float32)

tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.Dense(units=10, input_dim=2, activation='sigmoid'))
tf.model.add(tf.keras.layers.Dense(units=10, activation='sigmoid'))
tf.model.add(tf.keras.layers.Dense(units=10, activation='sigmoid'))
tf.model.add(tf.keras.layers.Dense(units=10, activation='sigmoid'))
tf.model.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

Vanishing Gradient 문제 때문에 optimizer로 SGD를 사용하면 학습이 잘 되지 않는다.
따라서 optimizer로 [Adam](http://mjgim.me/2018/01/22/adam.html) 을 사용. 

```python
tf.model.compile(
    loss='binary_crossentropy',
    optimizer=tf.optimizers.Adam(lr=0.1),
    metrics=['accuracy']
)
tf.model.summary()
```

```python
train_history = tf.model.fit(x_data, y_data, epochs=300)
```

```python
pred = tf.model.predict(x_data)
print('Prediction: \n', pred)

score = tf.model.evaluate(x_data, y_data)
print('Accuracy: ', score[1])
```
