## Multi Variable Linear Regression

```python
import tensorflow as tf
import numpy as np

x_data = [[73., 80., 75.],
         [93., 88., 93.],
         [89., 91., 90.],
         [96., 98., 100.],
         [73., 66., 70.]]

y_data = [[152.],
          [185.],
          [180.],
          [196.],
          [142.]]
```

Sequential 모델 생성후 Dense 계층 하나 쌓아준 후,
activation function 으로 linear를 설정해준다
* dense 계층 생성시 default activation function 으로 linear가 설정되어 있다.

```python
tf.model = tf.keras.Sequential()

tf.model.add(tf.keras.layers.Dense(units=1, input_dim=3))
# multi variables 이므로 input_dim 을 feature 개수만큼 지정해준다
tf.model.add(tf.keras.layers.Activation('linear')) # linear가 Activation의 기본값이므로 생략 가능
```

optimizer 와 loss function 설정

```python
sgd = tf.keras.optimizers.SGD(lr=1e-5)
tf.model.compile(loss='mse', optimizer = sgd)
tf.model.summary()
```

```python
train_history = tf.model.fit(x_data, y_data, epochs=100)
```

```python
y_predict = tf.model.predict(np.array([[72., 93., 90.]]))
print(y_predict)
```
