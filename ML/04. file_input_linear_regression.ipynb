## File Input Linear Regression

```python
import tensorflow as tf
import numpy as np

# np.loadtxt(파일경로, 구분자, 데이터형식)

data = np.loadtxt('./data/data-01-test-score.csv', delimiter=',', dtype=np.float32)
```

불러온 데이터를 확인  
column 1 ~ 3: train input  
column 4 : train output  

```python
print(data)
```

```python
x_data = data[:, 0:-1]
y_data = data[:, [-1]]
print(x_data, "\n shape : ", x_data.shape)
print()
print(y_data, "\n shape : ", y_data.shape)
```

```python
tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.Dense(units=1, input_dim=3, activation='linear'))
tf.model.summary()
```

```python
tf.model.compile(loss='mse', optimizer=tf.keras.optimizers.SGD(lr=1e-5))
train_history = tf.model.fit(x_data, y_data, epochs=2000)
```

학습한 모델로 점수 예측

```python
print("Your score will be ", tf.model.predict([[100, 70, 101]]))
print("Other scores will be ", tf.model.predict([[60, 70, 110], [90, 100, 80]]))
```
